```python
def fetch_coordinates(df, dim):
    coordinates = []
    for i in range(len(df)):
        for j in range(dim):
            coordinates.append(df.iloc[i,j])
    return coordinates
```

```python
def calculate_distance(s, cx, cy):
    distance = np.sqrt((s.loc['GA'] - cx)**2 + (s.loc['GF'] - cy)**2)
    return distance

def calculate_distance_assign_clusters(df, centroids):
    
    c0x, c0y, c1x, c1y, c2x, c2y, c3x, c3y = fetch_coordinates(centroids, 2)
    
    df['dist_centroid_0'] = df.apply(calculate_distance, args=(c0x, c0y), axis=1)
    df['dist_centroid_1'] = df.apply(calculate_distance, args=(c1x, c1y), axis=1)
    df['dist_centroid_2'] = df.apply(calculate_distance, args=(c2x, c2y), axis=1)
    df['dist_centroid_3'] = df.apply(calculate_distance, args=(c3x, c3y), axis=1)
    
    mask_0 = (df['dist_centroid_0'] < df['dist_centroid_2']) & (df['dist_centroid_0'] < df['dist_centroid_1']) & (df['dist_centroid_0'] < df['dist_centroid_3'])
    mask_1 = (df['dist_centroid_1'] < df['dist_centroid_2']) & (df['dist_centroid_1'] < df['dist_centroid_0']) & (df['dist_centroid_1'] < df['dist_centroid_3'])
    mask_2 = (df['dist_centroid_2'] < df['dist_centroid_0']) & (df['dist_centroid_2'] < df['dist_centroid_1']) & (df['dist_centroid_2'] < df['dist_centroid_3'])
    mask_3 = (df['dist_centroid_3'] < df['dist_centroid_2']) & (df['dist_centroid_3'] < df['dist_centroid_1']) & (df['dist_centroid_3'] < df['dist_centroid_0'])
    df.loc[mask_0, 'cluster'] = 0
    df.loc[mask_1, 'cluster'] = 1
    df.loc[mask_2, 'cluster'] = 2
    df.loc[mask_3, 'cluster'] = 3
    
    return df
    
    # Assign clusters
#     assign_c0 = np.where(df['dist_centroid_0'] == min(
#         df['dist_centroid_0'],df['dist_centroid_1'],df['dist_centroid_2'],df['dist_centroid_0']), 0, 3)
#     assign_c1 = np.where(df['dist_centroid_1'] == min(
#         df['dist_centroid_0'],df['dist_centroid_1'],df['dist_centroid_2'],df['dist_centroid_0']), 1, assign_c0)
#     df['clusters'] = np.where(df['dist_centroid_2'] == min(df['dist_centroid_0'],df['dist_centroid_1'],df['dist_centroid_2'],df['dist_centroid_0']), 2, np.where(df['dist_centroid_1'] == min(df['dist_centroid_0'],df['dist_centroid_1'],df['dist_centroid_2'],df['dist_centroid_0']), 1, np.where(df['dist_centroid_0'] == min(df['dist_centroid_0'],df['dist_centroid_1'],df['dist_centroid_2'],df['dist_centroid_0']), 0, 3)))
    

#     assign_c0 = np.where(df['dist_centroid_0'] == min(
#         df['dist_centroid_0'],df['dist_centroid_1'],df['dist_centroid_2'],df['dist_centroid_0']), 0, 3)
#     assign_c1 = np.where(pl23['dist_centroid_1'] == min(
#         df['dist_centroid_0'],df['dist_centroid_1'],df['dist_centroid_2'],df['dist_centroid_0']), 1, assign_c0)
#     df['clusters'] = np.where(df['dist_centroid_2'] == min(
#         df['dist_centroid_0'],df['dist_centroid_1'],df['dist_centroid_2'],df['dist_centroid_0']), 2, assign_c1)
```

```python
pl23 = calculate_distance_assign_clusters(pl23, centroids)
pl23['cluster'] = pl23['cluster'].astype(int)
```

```python
def n_iterations(df, c, loops):
    for i in range(loops):
        df = calculate_distance_assign_clusters(df, c)
        c = df.groupby(['cluster'])[['GA','GF']].mean().reset_index()
        c.drop('cluster', axis=1, inplace=True)
    return [df, c]
```

```python
pl23, new_centroids = n_iterations(pl23, new_centroids, 50000)
```
