# Seaborn Exercise

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
fig = plt.figure(figsize=(15,8))
sns.boxplot(data=tips, x='tip', y='time', hue='smoker')
plt.show(fig)
```

```python
fig = plt.figure(figsize=(15,8))
sns.violinplot(data=tips, x='tip', y='time', hue='smoker', split=True, palette='summer')
plt.show(fig)
```

```python
sns.countplot(data=tips, x='sex', hue='smoker')
```

```python
sns.pairplot(data=tips, hue='smoker')
```
