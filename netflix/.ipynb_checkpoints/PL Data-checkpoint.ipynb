```python
# Remove betting odds
pl23 = pl23.loc[:,'Div':'AR']
pl23.head()
```

```python
pl23.info()
```

```python
# Calculate goal difference for each match, positive indicates home team won negative indicates away 0 is a draw
gd = pl23['FTHG'] - pl23['FTAG']
print(gd)
```

```python
# Biggest home win gd
gd.max()
```

```python
# Biggest home defeat gd 
gd.min()
```

```python
# Average goal difference
home_w_draw = gd[gd >= 0]
print(home_w_draw)
```

```python
away_w = gd[gd < 0] * -1
print(away_w)
```

```python
# Average goal difference calculation
(home_w_draw.sum() + away_w.sum()) / gd.count()
```

```python
# Could do it by concatenating the arrays first
import numpy as np
np.concatenate((home_w_draw, away_w)).mean()
```

```python
# Pandas Index Alignment
# New df removing all the rows with null values for this particular stat
nn = pl_original.loc[pl_original['PC>2.5'].notnull() & pl_original['P>2.5'].notnull()]
# Change in odds
odds_change = nn['P>2.5'] - nn['PC>2.5']
# Reassign this to original df
pl_original['Change in >2.5 for P'] = odds_change
pl_original.head(100)
# Can see the 80th row has been assigned NaN as added a series with less elements than the df, used index alignment
# to add the new data correctly and assigned NaN to the other columns
```

```python
# Boolean operators
# All the games where a team scored at least 4 goals and won to nil
hw_4_0 = (pl23['FTHG'] >= 4) & (pl23['FTAG'] == 0)
aw_0_4 = (pl23['FTHG'] == 0) & (pl23['FTAG'] >= 4)
w4 = pl23.loc[aw_0_4 | hw_4_0]
```

```python
big_six = ['Arsenal', 'Liverpool', 'Man City', 'Man United', 'Chelsea', 'Tottenham']
# Games not involving big 6
nb6hw = (w4['FTHG'] < w4['FTAG']) | ((w4['HomeTeam'] != 'Liverpool') & (w4['HomeTeam'] != 'Arsenal') & (w4['HomeTeam'] != 'Tottenham') & (w4['HomeTeam'] != 'Man United') & (w4['HomeTeam'] != 'Man City'))
nb6aw = (w4['FTHG'] > w4['FTAG']) | ((w4['AwayTeam'] != 'Liverpool') & (w4['AwayTeam'] != 'Arsenal') & (w4['AwayTeam'] != 'Tottenham') & (w4['AwayTeam'] != 'Man United') & (w4['AwayTeam'] != 'Man City'))
w4[nb6hw & nb6aw]
```

```python
# Game in which chelsea scored the msot goals
pl23[pl23['AwayTeam']=='Chelsea'].sort_values('FTAG', ascending=False).head(1)
```

```python
# Biggest win for each time
teams = pl23['HomeTeam'].unique()
resultdf = pd.DataFrame()
resultdf.info()
print(teams)
```

```python
for team in teams:
    away_games = pl23[pl23['AwayTeam'] == team]
    home_games = pl23[pl23['HomeTeam'] == team]
    b_away_win = away_games['gd_home'].min()
    b_home_win = home_games['gd_home'].max()
    if b_home_win > b_away_win*-1:
        res = home_games.loc[home_games['gd_home'] == b_home_win, 'HomeTeam':'FTR'].head(1)
        resultdf = pd.concat([resultdf, res],axis=0,ignore_index=True)  
    else:
        res = away_games.loc[away_games['gd_home'] == b_away_win, 'HomeTeam':'FTR'].head(1)
        resultdf = pd.concat([resultdf, res],axis=0,ignore_index=True)  
resultdf.sort_values('AwayTeam', inplace=True)
resultdf.sort_values('HomeTeam', inplace=True)
resultdf.head(20)
```
