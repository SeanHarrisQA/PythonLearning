```python
away_w = gd[gd < 0] * -1
print(away_w)
```

```python
big_six = ['Arsenal', 'Liverpool', 'Man City', 'Man United', 'Chelsea', 'Tottenham']
# Games not involving big 6
nb6hw = (w4['FTHG'] < w4['FTAG']) | ((w4['HomeTeam'] != 'Liverpool') & (w4['HomeTeam'] != 'Arsenal') & (w4['HomeTeam'] != 'Tottenham') & (w4['HomeTeam'] != 'Man United') & (w4['HomeTeam'] != 'Man City'))
nb6aw = (w4['FTHG'] > w4['FTAG']) | ((w4['AwayTeam'] != 'Liverpool') & (w4['AwayTeam'] != 'Arsenal') & (w4['AwayTeam'] != 'Tottenham') & (w4['AwayTeam'] != 'Man United') & (w4['AwayTeam'] != 'Man City'))
w4[nb6hw & nb6aw]
```

```python
# Biggest win for each time
teams = pl23['HomeTeam'].unique()
resultdf = pd.DataFrame()
resultdf.info()
print(teams)
```

```python
for team in teams:
    away_games = pl23[pl23['AwayTeam'] == team]
    home_games = pl23[pl23['HomeTeam'] == team]
    b_away_win = away_games['gd_home'].min()
    b_home_win = home_games['gd_home'].max()
    if b_home_win > b_away_win*-1:
        res = home_games.loc[home_games['gd_home'] == b_home_win, 'HomeTeam':'FTR'].head(1)
        resultdf = pd.concat([resultdf, res],axis=0,ignore_index=True)  
    else:
        res = away_games.loc[away_games['gd_home'] == b_away_win, 'HomeTeam':'FTR'].head(1)
        resultdf = pd.concat([resultdf, res],axis=0,ignore_index=True)  
resultdf.sort_values('AwayTeam', inplace=True)
resultdf.sort_values('HomeTeam', inplace=True)
resultdf.head(20)
```

```python
# Using matplotlib

s = pl23['FTHG'] + pl23['FTAG']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(s.index, s.values)
```

```python
five_game_rolling_average = np.zeros(shape=(len(s.values) - 4))
```

```python
five_sum = sum(s.values[:5])
five_game_rolling_average[0] = five_sum / 5
print(five_sum)
```

```python
for i in range(5, len(s.values)):
    five_sum+=s.values[i]
    five_sum-=s.values[i-5]
    five_game_rolling_average[i - 4] = five_sum / 5


    
```

```python
# Five game rolling average of goals scored
plt.plot(range(len(five_game_rolling_average)),five_game_rolling_average)
plt.title('5 Game Rolling Average of Goals Scored')
plt.xlabel('Games')
plt.ylabel('Goals')
```
