```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
weights = pd.read_csv('7DAY Weight - Sheet1.csv')
```

```python
weights.head()
```

```python
core = weights[['Day','Weight']]
```

```python
core.sample(10)
```

```python
core.info()
```

```python
core['Day'] = core[['Day']].apply(lambda x: pd.to_datetime(x))
```

```python
core.info()
```

```python
core.set_index('Day', inplace=True)
```

```python
core.head()
```

```python
# Pass 0 to y to plot the index
core['Date'] = core.index
core.plot.scatter(x='Date', y=0, grid=True, figsize=(10,8), rot=90)
```

```python
core.loc[:,['Weight']].plot(grid=True, figsize=(10,8))
```

```python
core.loc[:,['rolling7']] = core['Weight'].rolling(window=7).mean()
```

```python
core.tail()
```

```python
core.loc[:,['rolling28']] = core['Weight'].rolling(window=28).mean()
```

```python
core.head(30)
```

```python
core.loc[:,['Weight']].plot(grid=True, figsize=(10,8))
core['rolling7'].plot(figsize=(10,8))
core['rolling28'].plot(figsize=(10,8))
# core.plot.scatter(x='Date', y=0, figsize=(10,8), rot=90)
plt.legend()
plt.show()
```

```python
core['rolling28'].min()
```

```python
core['rolling7'].min()
```

```python
core['Weight'].min()
```

```python
core['Weight_diff_7'] = core['rolling7'].diff(periods=7)
core['rwd7'] = core['Weight_diff_7'].rolling(window=7).mean()
```

```python
core
```

```python
core.loc[:,['Weight', 'Weight_diff_7','rwd7']].plot(grid=True, figsize=(10,8), secondary_y=['Weight_diff_7', 'rwd7'], mark_right=True)
core['rolling7'].plot(figsize=(10,8))
core['rolling28'].plot(figsize=(10,8))
# core['rwd7'].plot(figsize=(10,8))
# core.plot.scatter(x='Date', y=0, figsize=(10,8), rot=90)
plt.legend()
plt.show()
```
